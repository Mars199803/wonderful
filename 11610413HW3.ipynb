```python
#question 4

from tkinter import *
 
calculate = Tk()
calculate.geometry('400x400')
calculate.title('Calculator')
def f(root, side):
    aaa = Frame(root)
    aaa.pack(side = side, expand = YES, fill = BOTH)
    return aaa
def b(root, side, text, command = None):
    aaa = Button(root, text = text, command = command)
    aaa.pack(side = side, expand = YES, fill = BOTH)
    return aaa
class Calculator(Frame):
    def __init__(self):     
        Frame.__init__(self)       
        self.pack(expand = YES, fill = BOTH)
        self.master.title('Calculater')        
        show = StringVar()
        Entry(self, relief = SUNKEN,
              textvariable = show).pack(side = TOP, expand = YES,
                                           fill = BOTH)
        ceF = f(self, BOTTOM)
        b(ceF, LEFT, 'CE', lambda aaa = show:w.set(''))
        
        for key in('123', '456', '789', '+0.'):
            keyF = f(self, TOP)
            for char in key:
                b(keyF, LEFT, char, lambda aaa = show, c = char:aaa.set(aaa.get() + c))
        jjcc = f(self, TOP)
        for char in '-*/=':
            if char == '=':
                btn = b(jjcc, LEFT, char)
                btn.bind('<ButtonRelease - 1>', lambda e, s = self, aaa = show:s.calc(aaa), '+') 
            else:
                btn = b(jjcc, LEFT, char, lambda aaa = show, s = '%s' %char:aaa.set(aaa.get() + s))
    def calc(self, show):
        try:
            show.set(eval(show.get()))
        except:
            show.set("ERROR")
if __name__ == '__main__':
    Calculator().mainloop()
```

```python
#quesion 5
from tkinter import *           
import time
import random
ball_game=Tk()    
ball_game.title("Ball Game")
background=Canvas(ball_game,width=700,height=500,bg="pink",bd=0,highlightthickness = 0)
ball_game.resizable(0,0) 
background.pack() 
ball_game.update()   


class Ball: 
    def __init__(self,background,batten,color):
        self.background=background
        self.batten=batten
        self.id=background.create_oval(10,10,25,25,fill=color)
        self.background.move(self.id,200,100)    
        stat=[-3,-2,-1,1,2,3]   
        random.shuffle(stat)
        self.x=stat[0]
        self.y=-3
        self.background_height=self.background.winfo_height()  
        self.background_width=self.background.winfo_width()
        self.hit_bottom=False 
    def hit_batten(self, pos):     
        batten_pos = self.background.coords(self.batten.id )
        if pos[2]>= batten_pos[0] and pos[0]<= batten_pos[2]:
            if pos[3]>= batten_pos[1] and pos[3]<= batten_pos[3]:
                return True
        return False
    def draw(self): 
        self.background.move(self.id,self.x,self.y)
        pos=self.background.coords(self.id)
        if pos[1]<=0:  
            self.y=3
        if pos[3]>=self.background_height:
            self.hit_bottom=True
        if self.hit_batten(pos)==True:  
            self.y=-3
        if pos[0]<=0: 
            self.x=3
        if pos[2]>=self.background_width:
            self.x=-3
class Batten: 
    def __init__(self,background,color):
        self.background=background
        self.id=background.create_rectangle(0,0,150,10,fill=color)
        self.background.move(self.id,400,450)
        self.x=0
        self.background_width=self.background.winfo_width()
        self.background.bind_all("<KeyPress-Left>",self.LEFT) 
        self.background.bind_all("<KeyPress-Right>", self.RIGHT)
    def LEFT(self,event): 
        self.x=-5
    def RIGHT(self,event):
        self.x=5
    def draw(self):
        pos=self.background.coords(self.id)
        self.background.move(self.id, self.x, 0)
        if pos[0]<=0:
            self.x=0
        if pos[2]>=self.background_width:
            self.x=0
batten=Batten(background,"white")
ball=Ball(background,batten,"red")
while True: 
    if ball.hit_bottom==False:
        ball.draw()
        batten.draw()
    else:
        break
    ball_game.update_idletasks()
    ball_game.update()
    time.sleep(0.01)
```
